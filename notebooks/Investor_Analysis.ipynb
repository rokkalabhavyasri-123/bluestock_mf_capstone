```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    r"C:\Users\rokka\OneDrive\Desktop\bluestock_mf_capstone\data\raw\08_investor_transactions.csv"
)

df.head()
```

```python
df['age_group'].value_counts().plot(kind='bar')
plt.title("Age Group Distribution")
plt.show()
```

```python
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
df['city_tier'].value_counts().plot(kind='bar')
plt.title("City Tier Distribution")
plt.show()
```

```python
df['transaction_type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Transaction Type Distribution")
plt.ylabel("")
plt.show()
```

Key Findings:

• Highest transactions from ______ state
• Most investors belong to ______ age group
• SIP transactions dominate investor activity
• Tier-1/Tier-2 cities contribute majority of transactions
• Male/Female participation percentage observed

```python
df['annual_income_lakh'].plot(kind='hist', bins=10)
plt.title("Income Distribution")
plt.show()
```

```python
state_amount = df.groupby('state')['amount_inr'].sum().sort_values(ascending=False).head(10)

state_amount.plot(kind='bar')
plt.title("Top States by Investment Amount")
plt.show()
```

```python
pd.crosstab(
    df['city_tier'],
    df['transaction_type']
).plot(kind='bar', stacked=True)

plt.title("Transaction Type by City Tier")
plt.show()
```

Key Findings:

• SIP transactions dominate investor activity.
• Tier-1 cities contribute the highest transaction volume.
• Majority of investors belong to the most active age group identified in the analysis.
• Top states account for a significant share of total investments.
• Higher-income investors tend to contribute larger investment amounts.
